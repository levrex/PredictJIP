#### Create new features

```python

L_TJC_44 = [i for i in L_DAS_Joints_44 if 'Pijn_' in i] 
L_SJC_44 = [i for i in L_DAS_Joints_44 if 'Zwelling_' in i] 

# Count total swelling hand & foot
metadata['SJC_FOOT'] = metadata[L_Foot_Swelling_44].sum(axis=1)
metadata['SJC_HAND'] = metadata[L_Hand_Swelling_44].sum(axis=1)

# # Count total tenderness hand & foot
metadata['TJC_FOOT'] = metadata[L_Foot_Tender_44].sum(axis=1)
metadata['TJC_HAND'] = metadata[L_Hand_Tender_44].sum(axis=1)

# [not a feature for prediction] total jc 
metadata['JC_FOOT'] = metadata['SJC_FOOT'] + metadata['TJC_FOOT']
metadata['JC_HAND'] = metadata['SJC_HAND'] + metadata['TJC_HAND']

# Total swelling & tenderness
metadata['SJC44'] = metadata[L_SJC_44].sum(axis=1)
metadata['TJC44'] = metadata[L_TJC_44].sum(axis=1)

# Total count Small / Big
metadata['Big44'] = metadata[L_Big_Joints_44].sum(axis=1)
metadata['Small44'] = metadata[L_Small_Joints_44].sum(axis=1)

# Fraction hand versus foot
metadata['handFrac'] = metadata['JC_HAND'] / (metadata['JC_HAND'] + metadata['JC_FOOT'] + 1e-3)
# Fraction big versus small
metadata['bigFrac'] = metadata['Big44'] / (metadata['Small44'] + metadata['Big44'] + 1e-3)
```

## 2. Train KAN-encoder for ensemble 

We use KAN-encoder (need 'kan' module)

- See below example for optimizing KAN encoder (for detection of JIP-poly)

```python
#from kan import KAN
from sklearn.model_selection import train_test_split

from kan import *
from sklearn.metrics import confusion_matrix, plot_confusion_matrix
from sklearn.metrics import accuracy_score, precision_score, roc_auc_score
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
import numpy as np
#import xgboost as xgb
import optuna
import pickle
from sklearn.metrics import precision_recall_curve, average_precision_score
import matplotlib.pyplot as plt
import time
import seaborn as sn
from sklearn.metrics import confusion_matrix, plot_confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
import torch

optuna.logging.set_verbosity(optuna.logging.WARNING)

metadata['Poly_cluster'] = metadata['PhenoGraph_clusters']==3
TARGET2 = 'Poly_cluster'

def objective(trial):
    # Define label data
    # Import data
    X = metadata[cols_data].astype(float)#.values
    y = np.array(metadata[TARGET2].astype(float))
    # Bookmark all predictions
    y_pred = []
    
    # Apply 5 fold CV
    train_index, test_index = train_test_split(X.index, test_size=0.2, stratify=metadata[TARGET2].astype(float), random_state=trial._trial_id)
    print(train_index, test_index)
    X_train, X_test = metadata[cols_data].loc[train_index].values, metadata[cols_data].loc[test_index].values
    y_train, y_test = np.take(y, np.array(train_index)), np.take(y, np.array(test_index))

    dataset = {}
    dataset['train_input'] = torch.from_numpy(np.array(X_train)).float()
    dataset['test_input'] = torch.from_numpy(np.array(X_test)).float()
    dataset['train_label'] = torch.from_numpy(np.array(y_train)).view(-1, 1).float()
    dataset['test_label'] = torch.from_numpy(np.array(y_test)).view(-1, 1).float()

    param = {}
    param["width"] = [14,5,1]#, 
    param["grid"] = trial.suggest_int("grid", 1, 5)
    param["k"] = trial.suggest_int("k", 1, 5)
    param["lambda"] = trial.suggest_loguniform("lambda", 1e-5, 1.0)
    param["seed"] = 1234
    
    print(param)
    # create a KAN: 2D inputs, 1D output, and 5 hidden neurons. cubic spline (k=3), 5 grid intervals (grid=5).
    model = KAN(width=param['width'], grid=param['grid'], k=param['k']) # 2,3,1


    def train_acc():
        return torch.mean((torch.round(model(dataset['train_input'])[:,0]) == dataset['train_label'][:,0]).float())

    def test_acc():
        return torch.mean((torch.round(model(dataset['test_input'])[:,0]) == dataset['test_label'][:,0]).float())

    t0 = time.time()
    #results = model.train(dataset, opt="LBFGS", steps=20, metrics=(train_acc, test_acc));
    results = model.fit(dataset, opt="LBFGS", steps=20,  metrics=(train_acc, test_acc), lamb=param['lambda']) #lamb=0.001,

    print(results['train_acc'][-1], results['test_acc'][-1])
    t1 = time.time()
    print('Time for Training KAN model %s: %s' % (str(trial._trial_id), str(t1-t0)))

    print('trial id:', trial._trial_id)
    # accuracy_score
    return results['test_acc'][-1] # precision recall curve

#study = optuna.create_study(direction="minimize")
study = optuna.create_study(direction="maximize")
study.optimize(lambda trial: objective(trial), n_trials=10) # test # 2000 # -1 -> bootstrapping (take random 100 samples) , n_jobs=1
print('Best trial: %s' % study.best_trial.number)
print('Performance (auc): %s' % study.best_trial.value)
print('Corresponding Parameters: %s' % study.best_trial.params)

import plotly as py
# optuna.visualization.plot_intermediate_values(study)
fig = optuna.visualization.plot_optimization_history(study)
py.offline.plot(fig, filename='hyperparamtuning_optimization_kan_footcluster_noESR.html', auto_open=False)

#fig = optuna.visualization.plot_contour(study)
#py.offline.plot(fig, filename='hyperparamtuning_contour.html', auto_open=False)
```
